```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("customer_journey.csv")

print("Dataset loaded successfully.")
print("Shape:", df.shape)
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
duplicate_count = df.duplicated().sum()
print("Duplicate Rows:", duplicate_count)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns
```

```python
df = df.rename(columns={
    'sessionid': 'session_id',
    'userid': 'user_id',
    'timestamp': 'timestamp',
    'pagetype': 'page_type',
    'devicetype': 'device_type',
    'country': 'country',
    'referralsource': 'referral_source',
    'timeonpage_seconds': 'time_on_page_seconds',
    'itemsincart': 'items_in_cart',
    'purchased': 'purchased'})
df.columns
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
df['timestamp'].head()
```

```python
df['timestamp'].isnull().sum()
```

```python
df = df.sort_values(by=['session_id', 'timestamp']).reset_index(drop=True)
df.head(20)
```

```python
text_cols = ['page_type', 'device_type', 'country', 'referral_source']
for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.lower()
df.head()
```

```python
for col in ['page_type', 'device_type', 'country', 'referral_source', 'purchased']:
    print(f"\n--- {col.upper()} ---")
    print(df[col].value_counts())
```

```python
df['purchased'] = df['purchased'].astype(str).str.strip().str.lower()
df['purchased'] = df['purchased'].replace({
    'true': 1,
    'false': 0,
    'yes': 1,
    'no': 0,
    '1': 1,
    '0': 0})

df['purchased'] = df['purchased'].astype(int)
df['purchased'].value_counts()
```

```python
print("Negative time_on_page_seconds:", (df['time_on_page_seconds'] < 0).sum())
print("Negative items_in_cart:", (df['items_in_cart'] < 0).sum())
```

```python
df.loc[df['time_on_page_seconds'] < 0, 'time_on_page_seconds'] = np.nan
df.loc[df['items_in_cart'] < 0, 'items_in_cart'] = np.nan
df[['time_on_page_seconds', 'items_in_cart']].isnull().sum()
```

```python
df['date'] = df['timestamp'].dt.date
df['year'] = df['timestamp'].dt.year
df['month'] = df['timestamp'].dt.month
df['month_name'] = df['timestamp'].dt.month_name()
df['day'] = df['timestamp'].dt.day
df['day_name'] = df['timestamp'].dt.day_name()
df['hour'] = df['timestamp'].dt.hour

df.head()
```

```python
df.to_csv(r"E:\User_Funnel\customer_journey_clean.csv", index=False)

print("Cleaned dataset saved successfully.")
```

```python
session_summary = df.groupby('session_id').agg(
    user_id=('user_id', 'first'),
    session_start=('timestamp', 'min'),
    session_end=('timestamp', 'max'),
    total_pages_visited=('page_type', 'count'),
    total_time_spent_seconds=('time_on_page_seconds', 'sum'),
    max_items_in_cart=('items_in_cart', 'max'),
    purchased=('purchased', 'max'),
    device_type=('device_type', 'first'),
    country=('country', 'first'),
    referral_source=('referral_source', 'first')
).reset_index()

session_summary.head()
```

```python
session_summary['session_duration_minutes'] = (
    (session_summary['session_end'] - session_summary['session_start']).dt.total_seconds() / 60
)

session_summary.head()
```

```python
print("Session Summary Shape:", session_summary.shape)
session_summary.info()
```

```python
session_summary.to_csv(r"E:\User_Funnel\session_summary.csv", index=False)

print("Session summary saved successfully.")
```

```python
funnel_flags = df.groupby('session_id').agg(
    visited_home=('page_type', lambda x: int('home' in x.values)),
    visited_product_page=('page_type', lambda x: int('product_page' in x.values)),
    visited_cart=('page_type', lambda x: int('cart' in x.values)),
    visited_checkout=('page_type', lambda x: int('checkout' in x.values)),
    visited_confirmation=('page_type', lambda x: int('confirmation' in x.values)),
    purchased=('purchased', 'max')
).reset_index()

funnel_flags.head()
```

```python
session_master = session_summary.merge(funnel_flags, on=['session_id', 'purchased'], how='left')
session_master.head()
```

```python
session_master.to_csv(r"E:\User_Funnel\session_master.csv", index=False)

print("Session master table saved successfully.")
```

```python
print("Total Sessions:", session_master['session_id'].nunique())
print("Purchased Sessions:", session_master['purchased'].sum())
print("Conversion Rate %:", round(session_master['purchased'].mean() * 100, 2))
```

```python
print("Visited Home:", session_master['visited_home'].sum())
print("Visited Product Page:", session_master['visited_product_page'].sum())
print("Visited Cart:", session_master['visited_cart'].sum())
print("Visited Checkout:", session_master['visited_checkout'].sum())
print("Visited Confirmation:", session_master['visited_confirmation'].sum())
print("Purchased:", session_master['purchased'].sum())
```

```python
view_to_product = session_master['visited_product_page'].sum() / session_master['visited_home'].sum() * 100
product_to_cart = session_master['visited_cart'].sum() / session_master['visited_product_page'].sum() * 100
cart_to_checkout = session_master['visited_checkout'].sum() / session_master['visited_cart'].sum() * 100
checkout_to_confirmation = session_master['visited_confirmation'].sum() / session_master['visited_checkout'].sum() * 100
overall_conversion = session_master['purchased'].sum() / session_master['visited_home'].sum() * 100

print("Home → Product Page %:", round(view_to_product, 2))
print("Product Page → Cart %:", round(product_to_cart, 2))
print("Cart → Checkout %:", round(cart_to_checkout, 2))
print("Checkout → Confirmation %:", round(checkout_to_confirmation, 2))
print("Overall Conversion %:", round(overall_conversion, 2))
```
